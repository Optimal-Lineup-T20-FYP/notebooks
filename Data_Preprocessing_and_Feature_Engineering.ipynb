## Preprocessing ball data with respect to extras bowled in the previous ball

### bbl

```python
bbl['ball'].sort_values().unique()
```

```python
for i in range(1,len(bbl)):
  if(bbl['match_id'][i] == bbl['match_id'][i-1] and bbl['innings'][i] == bbl['innings'][i-1]):
    if (not (math.isnan(bbl['wides'][i-1]))) or (not math.isnan(bbl['noballs'][i-1]) or (bbl['wicket_type'][i] == 'run out')):
      bbl['ball'][i] = bbl['ball'][i-1]
    elif(math.floor(bbl['ball'][i]) == math.floor(bbl['ball'][i-1])):
      bbl['ball'][i] = round(bbl['ball'][i-1]+.1,1)
      if(round(bbl['ball'][i] - math.floor(bbl['ball'][i]),1) > 0.6):
        curi = i
        while(bbl['ball'][curi] > math.floor(bbl['ball'][i]) and not bbl['legbyes'][curi] > 0):
          bbl['ball'][curi] = round(bbl['ball'][curi]-.1,1)
          curi -= 1
```

#Feature Engineering

## bbl

```python
bbl['total_runs_scored'] = 0
bbl['wickets_remaining'] = 10
bbl['balls_remaining'] = 120
bbl["striker_score"] = 0
bbl["non_striker_score"] = 0
bbl['partnership'] = 0
```

```python
for i in range (1,len(bbl)):
  if bbl['ball'][i]!=0.1 or (bbl['ball'][i]==0.1 and bbl['ball'][i-1]==0.1):
    bbl['total_runs_scored'][i] = bbl['total_runs_scored'][i-1] + bbl['runs_off_bat'][i-1] + bbl['extras'][i-1]
    if isinstance(bbl['wicket_type'][i-1],float) and math.isnan(bbl['wicket_type'][i-1]):
      bbl['wickets_remaining'][i] = bbl['wickets_remaining'][i-1]
    else:
      bbl['wickets_remaining'][i] = bbl['wickets_remaining'][i-1] - 1

for i in range(0,len(bbl)):
  bbl['balls_remaining'][i] = (120 - (((bbl['ball'][i])*10)//10)*6 - (bbl['ball'][i]*10)%10) + 1

bat1 = ""
bat2 = ""
for i in range (1,len(bbl)):
  if bbl['ball'][i] == 0.1 and bbl['ball'][i-1]!=0.1:
    bbl['striker_score'][i] = 0
    bbl['non_striker_score'][i] = 0
    continue
  bat1 = bbl['striker'][i]
  bat2 = bbl['non_striker'][i]
  #if strike is retained, striker score += runs scored in previous ball, and non striker score remains the same
  if bat1==bbl['striker'][i-1] and bat2==bbl['non_striker'][i-1]:
    bbl['striker_score'][i] = bbl['striker_score'][i-1] + bbl['runs_off_bat'][i-1]
    bbl['non_striker_score'][i] = bbl['non_striker_score'][i-1]
  #if there is a change in strike - strikers score - non striker's score, non strikers score += runs scored in prev ball
  elif bat2==bbl['striker'][i-1] and bat1==bbl['non_striker'][i-1]:
    bbl['striker_score'][i] = bbl['non_striker_score'][i-1] 
    bbl['non_striker_score'][i] = bbl['striker_score'][i-1] + bbl['runs_off_bat'][i-1]
  #fall of a wicket
  else:
    playerout = bbl['player_dismissed'][i-1]
    #check if current striker is prev ball's striker - add runs off bat to striker, non striker - new batsman  - 0
    if bat1 == bbl['striker'][i-1]:
      bbl['striker_score'][i] = bbl['striker_score'][i-1] + bbl['runs_off_bat'][i-1]
      bbl['non_striker_score'][i] = 0
    #check if current striker is prev ball's non striker - add runs off bat to striker, non striker - new batsman  - 0
    elif bat1 == bbl['non_striker'][i-1]:
      bbl['striker_score'][i] = bbl['non_striker_score'][i-1]
      bbl['non_striker_score'][i] = 0
    else:
      bbl['striker_score'][i] = 0
      if bat2 == bbl['striker'][i-1]:
        bbl['non_striker_score'][i] = bbl['striker_score'][i-1] + bbl['runs_off_bat'][i-1]
      elif bat2 == bbl['non_striker'][i-1]:
        bbl['non_striker_score'][i] = bbl['non_striker_score'][i-1]

def isNaN(string):
  return string != string

for i in range (1,len(bbl)):
  if bbl['ball'][i] == 0.1 and bbl['ball'][i-1]!=0.1:
    continue
  #if player is dismissed in previous ball
  if not isNaN(bbl['player_dismissed'][i-1]):
    bbl['partnership'][i]=0
  else:
    bbl['partnership'][i] = bbl['partnership'][i-1] + bbl['runs_off_bat'][i-1] + bbl['extras'][i-1]
```

```python
for i in range (1,len(psl)):
  if psl['ball'][i]!=0.1 or (psl['ball'][i]==0.1 and psl['ball'][i-1]==0.1):
    psl['total_runs_scored'][i] = psl['total_runs_scored'][i-1] + psl['runs_off_bat'][i-1] + psl['extras'][i-1]
    if isinstance(psl['wicket_type'][i-1],float) and math.isnan(psl['wicket_type'][i-1]):
      psl['wickets_remaining'][i] = psl['wickets_remaining'][i-1]
    else:
      psl['wickets_remaining'][i] = psl['wickets_remaining'][i-1] - 1

for i in range(0,len(psl)):
  psl['balls_remaining'][i] = (120 - (((psl['ball'][i])*10)//10)*6 - (psl['ball'][i]*10)%10) + 1

bat1 = ""
bat2 = ""
for i in range (1,len(psl)):
  if psl['ball'][i] == 0.1 and psl['ball'][i-1]!=0.1:
    psl['striker_score'][i] = 0
    psl['non_striker_score'][i] = 0
    continue
  bat1 = psl['striker'][i]
  bat2 = psl['non_striker'][i]
  #if strike is retained, striker score += runs scored in previous ball, and non striker score remains the same
  if bat1==psl['striker'][i-1] and bat2==psl['non_striker'][i-1]:
    psl['striker_score'][i] = psl['striker_score'][i-1] + psl['runs_off_bat'][i-1]
    psl['non_striker_score'][i] = psl['non_striker_score'][i-1]
  #if there is a change in strike - strikers score - non striker's score, non strikers score += runs scored in prev ball
  elif bat2==psl['striker'][i-1] and bat1==psl['non_striker'][i-1]:
    psl['striker_score'][i] = psl['non_striker_score'][i-1] 
    psl['non_striker_score'][i] = psl['striker_score'][i-1] + psl['runs_off_bat'][i-1]
  #fall of a wicket
  else:
    playerout = psl['player_dismissed'][i-1]
    #check if current striker is prev ball's striker - add runs off bat to striker, non striker - new batsman  - 0
    if bat1 == psl['striker'][i-1]:
      psl['striker_score'][i] = psl['striker_score'][i-1] + psl['runs_off_bat'][i-1]
      psl['non_striker_score'][i] = 0
    #check if current striker is prev ball's non striker - add runs off bat to striker, non striker - new batsman  - 0
    elif bat1 == psl['non_striker'][i-1]:
      psl['striker_score'][i] = psl['non_striker_score'][i-1]
      psl['non_striker_score'][i] = 0
    else:
      psl['striker_score'][i] = 0
      if bat2 == psl['striker'][i-1]:
        psl['non_striker_score'][i] = psl['striker_score'][i-1] + psl['runs_off_bat'][i-1]
      elif bat2 == psl['non_striker'][i-1]:
        psl['non_striker_score'][i] = psl['non_striker_score'][i-1]

def isNaN(string):
  return string != string

for i in range (1,len(psl)):
  if psl['ball'][i] == 0.1 and psl['ball'][i-1]!=0.1:
    continue
  #if player is dismissed in previous ball
  if not isNaN(psl['player_dismissed'][i-1]):
    psl['partnership'][i]=0
  else:
    psl['partnership'][i] = psl['partnership'][i-1] + psl['runs_off_bat'][i-1] + psl['extras'][i-1]
```
